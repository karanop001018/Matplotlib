```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
blood_sugar = [113,85,90,150,149,88,93,115,135,80,77,82,129]
# plt.hist(blood_sugar,bins=3,rwidth=0.95) #rwidth is the space between the bars
# plt.hist(blood_sugar,bins=[80,100,125,150],rwidth=0.95,color='r',histtype="step") #rwidth is the space between the bars
plt.hist(blood_sugar,bins=[80,100,125,150],rwidth=0.95,color='orange')
```

```python
blood_sugar_men = [113,85,90,150,149,88,93,115,135,80,77,82,129]
blood_sugar_women = [67,98,89,120,133,150,84,69,89,79,120,112,100]
plt.xlabel("Blood Sugar Level")
plt.ylabel("Number of Patients")
plt.title("Blood Sugar Analysis")
plt.hist([blood_sugar_men, blood_sugar_women], bins=[80,100,125,150], rwidth=0.90, color=['blue','pink'],label=["Men","Women"])
# plt.hist([blood_sugar_men, blood_sugar_women], bins=[80,100,125,150], rwidth=0.85, color=['blue','pink'],label=["Men","Women"],orientation="horizontal")
plt.legend()
```
