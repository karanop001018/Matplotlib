```python
# Q: Write a program to plot a line chart to depict the changing weekly Tomato prices for 5 weeks. Give appropriate axes labels.
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
weeks = [1,2,3,4,5]
tomato_price = [30,50,60,80,45]
plt.plot(weeks,tomato_price, color='green',linestyle='-.')
plt.xlabel("Week")
plt.ylabel("Price of Tomato")
plt.title("Tomato price Analysis")
plt.legend()
```
