```python
# Sine and Cosine Graph
```

```python
import matplotlib.pyplot as plt
import numpy as np

angles = [0,30,45,60,90]
x = [i*(np.pi/180) for i in angles]
plt.plot(angles,np.sin(x), marker = "+", label = "sin")
plt.plot(angles,np.cos(x), marker = "^", label = "cos")
plt.xticks(angles)
plt.title("Sin and Cos graph")
plt.xlabel("Angles")
plt.ylabel("sin and cos values")
# plt.show()
plt.legend()
```

```python
angle = np.linspace(-np.pi,np.pi)

plt.plot(angle,np.sin(angle), marker = "+", label = "sin", linestyle="-.")
plt.plot(angle,np.cos(angle), marker = "^", label = "cos",linestyle="-.")

plt.title("Sin and Cos graph")
plt.xlabel("Angles")
plt.ylabel("sin and cos values")
# plt.show()
plt.legend()
```

```python
angle = np.arange(0,4*np.pi,0.5)

plt.plot(angle,np.sin(angle), marker = "+", label = "sin", color='red')
plt.plot(angle,np.cos(angle), marker = "^", label = "cos", color='green')

plt.title("Sin and Cos graph")
plt.xlabel("Angles")
plt.ylabel("sin and cos values")
# plt.show()
plt.legend()
```
