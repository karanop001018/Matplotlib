```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
exp_vals = [1400,600,300,410,250]
exp_labels = ["Home Rent","Food","Phone/Internet Bill","Car","Other Utilities"]
```

```python
plt.axis("equal")
# plt.pie(exp_vals,labels=exp_labels, autopct='%0.2f%%') # "%0.2f%%" : it means 2 numbers after the decimal point. If we don't want decimal just write "%0.0f%%"
# plt.pie(exp_vals,labels=exp_labels,radius=2)
plt.pie(exp_vals,labels=exp_labels, autopct='%0.2f%%', shadow=True, explode = [0,0.3,0,0,0.2]) # explode: it cuts out a specific piece of pie chart depending on given value
# plt.pie(exp_vals,labels=exp_labels, autopct='%0.2f%%', startangle=90)
plt.show()



#  To save pie chart
plt.savefig("piechart.pdf",bbox_inches="tight",pad_inches=2,transparent=True)
```
