```python
"""
Q: Sanaya is doing some research. She has a stored line of Pascal's triangle numbers as ar2 as shown below:
            ar2 = [1,7,21,35,21,7,1]
  She want to plot the sine, cosine and tangent values for the same array(ar2).
  (1) She wants cyan color for sine plot line, red color for cosine plot line and the black color for tangent plot line.
  (2) Also, the tangent line should be dashed.
"""
```

```python
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline
```

```python
ar2 = [1,7,21,35,21,7,1]
# sine
y1 = np.sin(ar2)
plt.plot(ar2,y1,color='cyan')

# cosine
y2 = np.cos(ar2)
plt.plot(ar2,y2,color='red')

# tangent
y3 = np.tan(ar2)
plt.plot(ar2,y3,color="black",linestyle="dashed")


plt.legend()
```

```python
'''
Q: Write a program to plot a bar chart from medals(Gold,Silver,Bronze) won by three countries:
              Australia = [80,75,59]
              India = [36,25,30]
              England = [45,46,45]
        Make sure that the bars are separately visible.
'''
```

```python
import matplotlib.pyplot as plt
import numpy as np

# data
countries = ['Australia', 'India', 'England']
gold = [80, 36, 45]
silver = [75, 25, 46]
bronze = [59, 30, 45]

# create bar chart
barWidth = 0.25
r1 = np.arange(len(gold))
r2 = [x + barWidth for x in r1]
r3 = [x + barWidth for x in r2]

plt.bar(r1, gold, color='gold', width=barWidth, edgecolor='grey', label='Gold')
plt.bar(r2, silver, color='silver', width=barWidth, edgecolor='grey', label='Silver')
plt.bar(r3, bronze, color='#CD853F', width=barWidth, edgecolor='grey', label='Bronze')

# adding labels
plt.xlabel('Countries', fontweight='bold', fontsize=15)
plt.ylabel('No. of Medals', fontweight='bold', fontsize=15)
plt.xticks([r + barWidth for r in range(len(gold))], countries)

plt.legend()
plt.show()
```

```python
'''
Q: Write a program to create a Pie Chart, which shows the percentage of amount collected by the different section of a school.
          Section = ['A','B','C','D','E']
          Amount = [8000,12000,15700,9800,10600]
'''
```

```python
import matplotlib.pyplot as plt

# data
sections = ['A', 'B', 'C', 'D', 'E']
amounts = [8000, 12000, 15700, 9800, 10600]

# create pie chart
plt.pie(amounts, labels=sections, autopct='%1.1f%%')

# adding title
plt.title('Percentage of Amount Collected by Different Sections')

plt.show()
```

```python
'''
Q: Given the set of Data :
Weight measurement for 16 small orders of French-Fries (in grams).
78 72 69 81 63 67 65 75
79 74 71 83 71 79 80 69
(i) Create a simple histogram from the above data
(ii) Create a horizontal histogram from the above data
(iii) Create a step type of histogram from the above data
(iv) Create a cumulative histogram from the above data
'''
```

```python
import matplotlib.pyplot as plt

# data
weights = [78, 72, 69, 81, 63, 67, 65, 75, 79, 74, 71, 83, 71, 79, 80, 69]

# create simple histogram
plt.figure(figsize=[10,8])
plt.subplot(221)
plt.hist(weights, bins=range(min(weights), max(weights) + 1), alpha=0.7, rwidth=0.85)
plt.title('Simple Histogram')

# create horizontal histogram
plt.subplot(222)
plt.hist(weights, bins=range(min(weights), max(weights) + 1), alpha=0.7, rwidth=0.85, orientation='horizontal')
plt.title('Horizontal Histogram')

# create step type of histogram
plt.subplot(223)
plt.hist(weights, bins=range(min(weights), max(weights) + 1), alpha=0.7, rwidth=0.85, histtype='step')
plt.title('Step Type Histogram')

# create cumulative histogram
plt.subplot(224)
plt.hist(weights, bins=range(min(weights), max(weights) + 1), alpha=0.7, rwidth=0.85, cumulative=True)
plt.title('Cumulative Histogram')

plt.tight_layout()
plt.show()
```
